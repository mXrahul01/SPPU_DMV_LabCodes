# Title : Analyzing Sales Performance by Region in a Retail Company

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Datasets/customer_shopping_data.csv") 
df.head()
```

```python
df.tail()
```

##### count of records grouped by region/branch of the mall

```python
df.groupby("shopping_mall").count()
```

##### count of records grouped by the product categories

```python
 df.groupby("category").count()
```

##### total sales for each mall branch

```python
branch_sales = df.groupby("shopping_mall").sum(numeric_only=True)
branch_sales
```

##### total sales for each category of product

```python
category_sales = df.groupby("category").sum(numeric_only=True)
category_sales
```

##### Top Performin Mall Branches

```python
branch_sales.sort_values(by = "price", ascending = False)
```

##### top selling categories/Products

```python
category_sales.sort_values(by = "price", ascending = False)
```

##### to get total sales for each combination of branch and product_category

```python
combined_branch_category_sales = df.groupby(["shopping_mall", "category"]).sum(numeric_only=True)
combined_branch_category_sales
```

##### pie chart for sales by branch

```python
plt.pie(branch_sales["price"], labels = branch_sales.index) 
plt.show()
```

##### pie chart for sales by product category

```python
plt.pie(category_sales["price"], labels = category_sales.index) 
plt.show()
```

```python
combined_pivot = df.pivot_table(index="shopping_mall", columns="category", values="price", aggfunc="sum") 
```

```python
combined_pivot
```

##### grouped bar chart for sales of different categories at different branches

```python
combined_pivot.plot(kind="bar", figsize=(10, 6)) 
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



shopping_mall = df["shopping_mall"]  # Branches
category = df["category"]            # Categories
price = df["price"]                  # Prices of products


# Use pd.Categorical to handle categorical data
mall_codes = pd.Categorical(shopping_mall).codes
category_codes = pd.Categorical(category).codes


fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')


ax.scatter(mall_codes, category_codes, price, c='r', marker='o')


ax.set_xlabel('Shopping Mall (Branch)')
ax.set_ylabel('Category')
ax.set_zlabel('Price')


ax.set_xticks(range(len(pd.Categorical(shopping_mall).categories)))
ax.set_xticklabels(pd.Categorical(shopping_mall).categories)

ax.set_yticks(range(len(pd.Categorical(category).categories)))
ax.set_yticklabels(pd.Categorical(category).categories)


plt.title("3D Plot of Shopping Mall, Category, and Price")
plt.show()
```
