```python
unified_data=pd.concat([csv,ed], ignore_index=True)
```

```python
unified_data.head()
```

```python
unified_data.shape
```

```python
total_sales= unified_data['SALES'].sum()
print("Total Sales:",total_sales)
```

```python
product_category_distribution = unified_data['PRODUCTLINE'].value_counts()
print(product_category_distribution)
```

```python
category_sales=unified_data.groupby('ORDERNUMBER')['SALES'].mean()
```

```python
category_sales
```

```python
category_counts = unified_data['SALES'].value_counts()
```

##### Bar Plots

```python
category_counts.plot(kind='bar')
plt.title('Product Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
unified_data.info()
```

#### Pie Plot

```python
status_counts = unified_data['STATUS'].value_counts()

plt.figure(figsize=(8, 6)) 
plt.pie(status_counts, labels=status_counts.index,  startangle=140)
plt.title('Distribution of Order Status')

plt.show()
```

```python
import seaborn as sns
```

#### BoxPlot

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='STATUS', y='SALES', data=unified_data)

plt.xlabel('Order Status')
plt.ylabel('Sales')
plt.title('Box Plot of Sales by Order Status')

plt.show()
```
